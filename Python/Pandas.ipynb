```python
dia_df.count() # no of entries against each column
```

# Manipulating a Df

```python
import pandas as pd
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing() #dataset
clf_df=pd.DataFrame(housing.data, columns=housing.feature_names)
clf_df.head()
```

```python
clf_df['Price']=housing.target # price column will be added of same rows
clf_df.head()
```

```python
clf_df.drop(index=2, axis=0) # row with index 2 will be deleted and return a new df
```

```python
clf_df.drop(columns='HouseAge', axis=1) # houseage col will be deleted and return a new df
```
